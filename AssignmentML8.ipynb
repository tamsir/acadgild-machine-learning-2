```python
import nltk
import urllib.request
from bs4 import BeautifulSoup
import re
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
response = urllib.request.urlopen('http://php.net/')
html = response.read()
soup = BeautifulSoup(html,"html5lib")
```

```python
print(html)
```

```python
print(soup)
```

```python
# For example, If we wan't to see the title of the page
print(soup.title)
print(soup.title.string)
```

```python
# if we wan't to get only the text

Alltext = soup.get_text()
print(Alltext)
```

```python
# From the text we use the regex to get the word list
WordList = re.findall('\w+', Alltext)
print(WordList)
```

```python
# Convert all the word to lower case
WordLower = []
for text in WordList:
    WordLower.append(text.lower())
print(WordLower)
```

```python
# Delete all the stop words in WordLower
import nltk
nltk.download('stopwords')
```

```python
stopWords = nltk.corpus.stopwords.words('english')
print(stopWords)
```

```python
print("Total including stop words:", len(WordLower))
clean_words = []
for w in WordLower:
    if w not in stopWords:
        clean_words.append(w)
    
print("Length after stop words:", len(clean_words) )
print("Total removed stop words:", len(WordLower)-len(clean_words) )
```

```python
print(clean_words)
```

```python
counts  = nltk.FreqDist(clean_words)
vocab   = len(counts.keys())
words   = sum(counts.values())
lexdiv  = float(words) / float(vocab)
print("Text has %i types and %i tokens for a lexical diversity of %0.3f" % (vocab, words, lexdiv))
```

```python
print(counts.most_common(100))
```

```python
print(counts.max())
```

```python
print(counts.hapaxes()[0:20])
```
